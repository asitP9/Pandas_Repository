```python
orders=pd.read_csv("order.csv")
```

```python
orders["price"]=orders.item_price.str[1:]
```

```python
orders["price"] = orders["price"].astype("float64")
```

```python
orders.dtypes
```

```python
orders.drop('item_price', inplace=True, axis=1)
```

```python
orders.dtypes
```

```python
# Calculate the sum of all the prices of order_id 1
orders[orders.order_id==1].price.sum()
```

```python
orders.groupby('order_id').price.sum().head()
```

```python
orders.groupby('order_id').price.agg(["sum", "count"]).head()
```

## COMBINE THE AGGREGATION OF THE OUTPUT WITH A DATAFRAME

```python
total_price=orders.groupby('order_id').price.transform('sum')
total_price
```

```python
orders['total_price']=total_price
```

```python
orders
```

```python
orders["percent_of_total"]=orders['price']/orders['total_price']
```
